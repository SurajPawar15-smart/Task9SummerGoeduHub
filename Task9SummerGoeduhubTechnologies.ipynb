***Label Encoding***<br>
It is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering. Let's see how to implement label encoding in Python using the scikit-learn library and also understand the challenges with label encoding.

```python
from sklearn.preprocessing import LabelEncoder
lab_encoding = LabelEncoder()
```

***its apply on the department as total 7 over here.***

```python
df.Department = lab_encoding.fit_transform(df.Department)
df.Department.unique()
```

```python
df.salary = lab_encoding.fit_transform(df.salary)
df.salary.unique()
```

***its apply on the salary as they are categorical values such as (low ,medium and high).***

***from X rows its remove the Sales value and on y in one Sales single feature present as(dependent and independent features).***

```python
X = df.drop(['salary'], axis=1)
y = df['salary']
```

***Splitting the dataset into Training and Testing set***

```python
#Splitting the dataset into Training-set and Testing-set
import sklearn
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
print("Shape of X_train",X_train.shape)
print("Shape of X_test",X_test.shape)
print("Shape of y_train",y_train.shape)
print("Shape of y_test",y_test.shape)
```

***Fitting Simple Logistic Regression to the Training Dataset***

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predicting the values (Scores)
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mse
```
